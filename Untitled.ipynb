```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
datasets = ['StreamSpot', 'Wget', 'Trace', 'Theia', 'Cadets']
fine_sizes = [3, 16, 11, 3, 6]  # KB单位，Wget的4.4MB转为4400KB
magic_sizes = [397, 172, 473, 259, 297]  # KB单位，StreamSpot和Wget的4.7MB、4.6MB转为4700KB、4600KB

# 设置柱状图
x = np.arange(len(datasets))  # 横轴位置
width = 0.35  # 柱子宽度

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, fine_sizes, width, label='FINE', color='skyblue')
bars2 = ax.bar(x + width/2, magic_sizes, width, label='MAGIC', color='lightcoral')

# 添加标签和标题
ax.set_xlabel('Dataset')
ax.set_ylabel('Training Time (S)')
ax.set_title('FINE vs MAGIC on Training Time ')
ax.set_xticks(x)
ax.set_xticklabels(datasets)
ax.legend()

# 在柱子上方显示具体数值
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 50, f'{int(yval)}KB', ha='center', va='bottom')

for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 50, f'{int(yval)}KB', ha='center', va='bottom')

# 调整y轴范围以避免标签重叠
ax.set_ylim(0, max(max(fine_sizes), max(magic_sizes)) + 500)

# 显示图形
plt.tight_layout()
plt.savefig('.jpg', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 数据
datasets = ['streamspot', 'wget', 'trace', 'theia', 'cadets']
train_times = [3, 16, 11, 3, 6]
test_times = [3, 8, 607, 196, 184]
# -------- 保存测试时间图 --------
plt.figure(figsize=(6, 5))
bars = plt.bar(datasets, test_times, color='salmon')
plt.title('Testing Time')
plt.xlabel('Dataset')
plt.ylabel('Time (s)')
plt.ylim(0, 650)

# 添加数值标签
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.5, f'{height}s', 
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig('testing_time.jpg', dpi=300)
plt.close()
```

```python
import matplotlib.pyplot as plt

# 模型与时间
models = ['FINE', 'MAGIC']
train_times = [11, 229]
test_times = [607, 830]

# 设置图表
bar_width = 0.35
x = range(len(models))

plt.figure(figsize=(8, 6))

# 绘制训练和测试时间
plt.bar(x, train_times, width=bar_width, label='Train Time', color='skyblue')
plt.bar([i + bar_width for i in x], test_times, width=bar_width, label='Test Time', color='salmon')

# 添加数值标签
for i in range(len(models)):
    plt.text(i, train_times[i] + 10, f'{train_times[i]}s', ha='center')
    plt.text(i + bar_width, test_times[i] + 10, f'{test_times[i]}s', ha='center')

# 设置标签与标题
plt.xlabel('Model')
plt.ylabel('Time (s)')
plt.title('FINE vs MAGIC on trace Dataset')
plt.xticks([i + bar_width / 2 for i in x], models)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 0.97), ncol=2)
plt.ylim(0, max(test_times) + 100)

plt.tight_layout()
plt.savefig('fine_vs_magic_trace.jpg', dpi=300)  # 可选：保存为图像文件
plt.show()
```
